```python
df2021=df[df['year']==2021]  #Fetching only records where Year=2021 for the first three tasks
df2021.drop(['year'],axis=1,inplace=True)  #dropping the year columns for the first three task
```

# TASK-1 

# FREQUENCY DISTRIBUTION OF THE VARIABLES

FOR INDEPENDENT NUMERICAL VARIABLE : Using Matplotlib and Seaborn Library(for better and detailed graph)

```python
sns.set(font_scale=1.2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

dx = sns.histplot(df2021['total.fertility.rate'],stat='density', ax=axes[0, 0])
dx.set_xlabel('(a) total.fertility.rate')
cx=sns.histplot(df2021['life.expectancy.both.sexes'],stat='density', ax=axes[0, 1])
cx.set_xlabel('(b) life.expectancy.both.sexes')
ax=sns.histplot(df2021['life.expectancy.females'],stat='density', ax=axes[1, 1])
ax.set_xlabel('(c) life.expectancy.females')
bx=sns.histplot(df2021['life.expectancy.males'],stat='density', ax=axes[1, 0])
bx.set_xlabel('(d) life.expectancy.males')

cx.set(ylim=(0.00, 0.08),xlim=(50,100))
ax.set(ylim=(0.00, 0.08),xlim=(50,100))
bx.set(ylim=(0.00, 0.08),xlim=(50,100))

plt.savefig('Figure 1.pdf')
```

# DIFFERENCES BETWEEN THE SEXES

```python
a=plt.scatter(df2021['life.expectancy.both.sexes'], df2021['life.expectancy.females'], color='r')
b=plt.scatter(df2021['life.expectancy.both.sexes'], df2021['life.expectancy.males'], color='b')
plt.xlabel('Life Expectancy Range',size=12)
plt.ylabel('Life Expectancy',size=12)
plt.title('Differences between the sexes')
plt.legend((a,b),('Females','Males'))
plt.savefig('Figure 2.pdf')
plt.show()
```

# TASK-3

# Comparison of variability within and among sub-region

```python
sns.set(rc={'figure.figsize':(8,8)})
```

```python
my_pal = sns.color_palette("bright")
```

```python
ax = sns.boxplot(data=(df2021.sort_values(["region"])),x='total.fertility.rate', y='subregion', hue='region', palette=my_pal,
                 dodge= False,width=0.3,linewidth=2.0)
ax.set(xlabel='total.fertility.rate', ylabel='subregion')
plt.subplots_adjust(left=0.25)
plt.savefig('Figure 4.pdf')
```

```python
ax = sns.boxplot(data=(df2021.sort_values(["region"])),x='life.expectancy.both.sexes', y='subregion', hue='region', palette=my_pal,
                dodge= False,width=0.3,linewidth=2.0)
ax.set(xlabel='Life Expectancy at Birth : Both Sexes', ylabel='subregion')
plt.subplots_adjust(left=0.25)
plt.savefig('Figure 6.pdf')
```

```python
ax = sns.boxplot(data=(df2021.sort_values(["region"])),x='life.expectancy.males', y='subregion', hue='region', palette=my_pal,
                 dodge= False,width=0.3,linewidth=2.0)
ax.set(xlabel='life.expectancy.males', ylabel='subregion')
plt.subplots_adjust(left=0.25)
plt.savefig('Figure 7.pdf')
```

```python
ax = sns.boxplot(data=(df2021.sort_values(["region"])),x='life.expectancy.males', y='subregion', hue='region', palette=my_pal,
                 dodge= False,width=0.3,linewidth=2.0)
ax.set(xlabel='life.expectancy.females', ylabel='subregion')
plt.subplots_adjust(left=0.25)
plt.savefig('Figure 8.pdf')
```
